```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
# Создаём сэмпл (взаимосвязь веса, роста, индекса BMI и длины волос (последнее - "лишний" признак))
n_samples = 1000

height = np.random.randint(158, 185, n_samples)+2
hair_length = np.random.randint(20, 60, n_samples)+3
bmi = np.random.randint(16, 32, n_samples)

weight = (height - 100)*bmi/20

data = pd.DataFrame({'height': height, 'hair_length': hair_length, 'bmi': bmi, 'weight': weight})
data.head(5)
```

```python
from sklearn.metrics import mean_absolute_error

X = data[['height', 'hair_length', 'bmi']]
y = data['weight']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['height', 'hair_length', 'bmi']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
X = data[['height', 'bmi']]
y = data['weight']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['height', 'bmi']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
data['new'] = data['weight'] / data['bmi']
data.head(5)
```

```python
X = data[['new']]
y = data['weight']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['new']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
